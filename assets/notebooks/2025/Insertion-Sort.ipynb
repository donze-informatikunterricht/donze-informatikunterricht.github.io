## 6. Analyse und Optimierung

Wenn eine Liste sortiert werden soll, gibt es drei mögliche Ausgangslagen:
  * **bestmögliche** Ausgangslage («best case»):\
    *bereits korrekt sortierte Liste*
  * **ungünstigste** Ausgangslage («worst case»):\
    *verkehrt herum sortierte Liste*
  * **mittelmässige** Ausgangslage («average case»):\
    *Liste mit zufällig angeordneten Elementen*
  
Anschliessend werden Sie den Algorithmus für die drei Ausgangslagen analysieren.

Sie können die folgende Zelle verwenden, um Ihre Listen zu generieren.

```python
import random                                     # nötig zum Erstellen der zufällig sortierten Liste

anzahl_elemente = 10 # <--- Hier können Sie die Anzahl Elemente steuern

aufsteigend = list(range(1, anzahl_elemente + 1)) # Bereich (kennen Sie)
absteigend  = aufsteigend[::-1]                   # Teilbereich der Liste (von hinten nach vorne mit Schrittweite -1)
zufaellig   = aufsteigend[:]                      # «Kopie von aufsteigend» machen und in zufaellig speichern
random.shuffle(zufaellig)

print(aufsteigend)
print(absteigend)
print(zufaellig)
```
